```python
#日期是否在考勤范围
def in_time_span(time):
    time = datetime.strptime(time,"%Y-%m-%d")
    current_time = datetime.now()
    month = 1 #current_time.month
    day = time.day
#     print(day)
#     print(month)
    last_month = month-1 if month-1>0 else 12
    if last_month==12:
        start_time = str(current_time.year-1)+'-'+str(last_month)+'-'+str(26)
    else:
        start_time = str(current_time.year)+'-'+str(last_month)+'-'+str(26)
    end_time = str(current_time.year)+'-'+str(month)+'-'+str(25)
    start_time = parse(start_time)
#     print(start_time)
    end_time = parse(end_time)
#     print(end_time)
    if time.month ==12 and time.day==26:
        return True
    elif time.month ==1 and time.day==25:
        return True
    else:
        return time>start_time and time<end_time
time = "2021-1-03"
in_time_span(time)
```

```python
#判断是否为周末

# def if_workday(day_str, separator="-"):
#     """
#     if a day is workday
#     :param day_str: string of a day
#     :param separator: separator of year, month and day, default is empty
#     :return: True: is workday; False: not workday
#     """
#     spec = "%Y" + separator + "%m" + separator + "%d"
#     day = datetime.strptime(day_str, spec).date()
#     # Monday == 0 ... Sunday == 6
#     if day.weekday() in [0, 1, 2, 3, 4]:
#         return True
#     else:
#         return False


def weekend(day_str, separator="-"):
    """
    if a day is weekend
    :param day_str: string of a day
    :param separator: separator of year, month and day, default is empty
    :return: True: is weekend; False: not weekend
    """
    spec = "%Y" + separator + "%m" + separator + "%d"
    day = datetime.datetime.strptime(day_str, spec).date()
    # Monday == 0 ... Sunday == 6
    return day.weekday()+1
```

```python
#验证日期格式
def verify_date_str_lawyer(datetime_str):
    try:        
        datetime.datetime.strptime(datetime_str, '%Y-%m-%d %H:%M')        
        return '1'   
    except ValueError:        
        return '2'
```

```python
#计算相隔多少天，但还要根据中间是否为周末
import datetime
def cross_day(start,end):
    if verify_date_str_lawyer(start)=='1' and verify_date_str_lawyer(end)=='1':
        start_time = datetime.datetime.strptime(start,"%Y-%m-%d %H:%M")
        end_time = datetime.datetime.strptime(end,"%Y-%m-%d %H:%M")
    else:
        start_time = datetime.datetime.strptime(start,"%Y-%m-%d")
        end_time = datetime.datetime.strptime(end,"%Y-%m-%d")
    return (end_time-start_time).days
starttime = "2021-01-13"
endtime = "2021-01-15"
cross_day(starttime,endtime)
```

```python
#计算下几天,下一天开始时间为9点
def next_day(time,daynums):
    today = datetime.datetime.strptime(time,"%Y-%m-%d %H:%M")
    flag =1#假定下一天是周末
    while(flag):
        
        if verify_date_str_lawyer(time)=='1':
            next_day = (today + datetime.timedelta(days=daynums)).strftime("%Y-%m-%d %H:%M")
        else:
            next_day = (today + datetime.timedelta(days=daynums)).strftime("%Y-%m-%d")
        if weekend(next_day.split(' ')[0])==6 or weekend(next_day.split(' ')[0])==7:
            flag=1
            daynums=daynums+1
        else:
            flag=0
    
    return next_day

def next_day_sx(time,daynums):
    today = datetime.datetime.strptime(time,"%Y-%m-%d %H:%M")
    flag =1#假定下一天是周末
    while(flag):
        
        if verify_date_str_lawyer(time)=='1':
            next_day = (today + datetime.timedelta(days=daynums)).strftime("%Y-%m-%d %H:%M")
        else:
            next_day = (today + datetime.timedelta(days=daynums)).strftime("%Y-%m-%d")
        if weekend(next_day.split(' ')[0])==7:
            flag=1
            daynums=daynums+1
        else:
            flag=0
    
    return next_day
next_day_sx("2021-01-15 09:00",1).split(' ')[0]
```

```python
#处理时间跨度，合并时间
#返回该员工这次请假的总时长，以及详细时间跨度，精确到每一天,需要返回一个合并时间
def process_time_span(starttime,endtime,durationtime,flag):
    """durationtime float 
        starttime str
        endtime str
        flag =0 :中华，flag=1 善心， int
    """
    timeformat = "%Y-%m-%d %H:%M"
    if verify_date_str_lawyer(starttime)=='1' and verify_date_str_lawyer(endtime)=='1':
        time1 = datetime.datetime.strptime(starttime,timeformat)
        time2 = datetime.datetime.strptime(endtime,timeformat)
    else:
        starttime = starttime+' '+'09:00'
        endtime  = endtime+' '+'18:00'
        time1 = datetime.datetime.strptime(starttime,timeformat)
        time2 = datetime.datetime.strptime(endtime,timeformat)
    total_hours = 0
    concat_time = []
    #一天内（8小时）
    if durationtime<= 8:
        if verify_date_str_lawyer(starttime)=='1' and verify_date_str_lawyer(endtime)=='1':
            #处理'.0'
            temp_int = int(durationtime)
            if durationtime-temp_int==0:
                temp_str = str(durationtime).split('.0')[0]
                temp = starttime+'-'+endtime.split(' ')[1]+' '+temp_str
            else:
                temp = starttime+'-'+endtime.split(' ')[1]+' '+str(durationtime)
            concat_time.append(temp)#这里要处理
    else:
        count= int(durationtime)//8
#         temp_total = 0
#         if durationtime%8==0:
        #先把整除的情况先算好，然后处理不整除的情况
        for i in range(0,count):
            temp=''
            if flag==0:#中华员工
                temp = next_day(starttime,i)#下i天的起始时间，9点开始
            else:
                temp = next_day_sx(starttime,i)
            temp = temp+'-'+'18:00'+' '+str(8)#
            concat_time.append(temp)#/////
#                 temp_total=temp_total+8
        if count*8<durationtime:#说明还要额外一天的部分时间进行补充
            left_hours= durationtime-count*8
            
            end_hour = 9+left_hours
            if end_hour>12:#超过12点，则要加一个小时
                end_hour=end_hour+1
                if len(str(end_hour).split('.'))>1:
                    temp = str(format(float('0.'+str(end_hour).split('.')[1])*60,'.2f')).split('.')[1]#多出来的小数部分转化为分钟数
                    end_hour = str(end_hour).split('.')[0]+':'+ temp
                else:#不含小数
                    end_hour = str(end_hour)+':'+'00'
            else:#未超过12点
                if len(str(end_hour).split('.'))>1:
                    temp = str(float('0.'+str(end_hour).split('.')[1])*60).split('.')[1]#多出来的小数部分转化为分钟数
                    end_hour = str(end_hour).split('.')[0]+':'+ temp
                else:#不含小数
                    end_hour = str(end_hour)+':'+'00'
            if flag==0:#中华员工
                temp = next_day(starttime,count)
            else:#善心体检
                temp = next_day_sx(starttime,count)
            if left_hours - int(left_hours)==0:
                temp_str = str(left_hours).split('.0')[0]
                temp = temp+'-'+end_hour+' '+str(temp_str)#
            else:
                temp = temp+'-'+end_hour+' '+str(left_hours)
            concat_time.append(temp)#///////
                    
                
    return concat_time

starttime = "2021-01-14 09:00"
endtime = "2021-01-18 15:00"
duration_time =32
print(process_time_span(starttime,endtime,duration_time,1))
        
        
    
    
```
